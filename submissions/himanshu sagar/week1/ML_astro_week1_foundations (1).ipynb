```python
x = torch.randn(3,64,64).cuda()

print(x[0].shape)
print(x[:,0,:].shape)
print(x[:,:,-1].shape)

```

```python
x= torch.randn(3,64,64).cuda()

flat = x.view(-1)

print("original=", x.shape)
print("Flattened=", flat.shape)
```

```python
x = torch.randn(3,64,64).cuda()

flat2 = x.reshape(3,-1)

print("original=", x.shape)
print("reshaped=", flat2.shape)
```
